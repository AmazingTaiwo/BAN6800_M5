```python
# =======================================================
# BAN6800  Module 4: Business Analytics Model Development
# Project: Databricks-Enabled Procurement Analytics
# Author: Taiwo Babalola
# Dataset: abc_dw_gl_pr_po_kpi.csv (Gold Layer)
# =======================================================

# -----------------------------
# 1. IMPORT LIBRARIES
# -----------------------------
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import display

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    RocCurveDisplay,
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

# NEW: SMOTE for balancing
from imblearn.over_sampling import SMOTE

# General display / plotting config
pd.set_option("display.max_columns", None)
plt.rcParams["figure.figsize"] = (8, 5)

# Create outputs folder for Excel + PNG
os.makedirs("outputs", exist_ok=True)


# =======================================================
# 2. LOAD DATA
# =======================================================
df = pd.read_csv("abc_dw_gl_pr_po_kpi.csv")

print("Dataset shape:", df.shape)
display(df.head())


# =======================================================
# 3. LIGHT EDA: STRUCTURE, MISSING VALUES, KPIs
# =======================================================
print("\n=== Data Info ===")
df.info()

key_cols_preview = [
    "pr_to_po_ageing",
    "pr_approval_ageing",
    "po_approval_ageing",
    "sla_breach_flag",
    "pr_cycle_sla_breach_flag",
    "po_cycle_sla_breach_flag",
    "record_type",
    "pr_companycode",
    "po_companycode",
    "pr_plant",
    "po_plant",
    "po_purchasingdoctypedesc",
    "po_countrykey"
]
existing_key_cols = [c for c in key_cols_preview if c in df.columns]

print("\n=== Sample of Key Fields ===")
if existing_key_cols:
    display(df[existing_key_cols].head())

numeric_kpi_cols = [
    c for c in ["pr_to_po_ageing", "pr_approval_ageing", "po_approval_ageing"]
    if c in df.columns
]

if numeric_kpi_cols:
    print("\n=== Descriptive Stats for Core Numeric KPIs ===")
    display(df[numeric_kpi_cols].describe())

print("\n=== Top 20 Columns by Missing Values ===")
missing_counts = df.isna().sum().sort_values(ascending=False)
display(missing_counts.head(20))


# =======================================================
# 4. CLASSIFICATION MODEL  SLA BREACH PREDICTION
# =======================================================

# -------------------------------------------------------
# 4.1 Prepare Target (SLA Breach)
# -------------------------------------------------------
clf_df = df[df["sla_breach_flag"].isin(["YES", "NO"])].copy()

# Binary target: 1 = SLA Breach, 0 = No Breach
clf_df["sla_breach_flag_bin"] = np.where(clf_df["sla_breach_flag"] == "YES", 1, 0)

print("\n=== SLA Breach Target Distribution (Binary) ===")
target_dist = clf_df["sla_breach_flag_bin"].value_counts(normalize=True).rename("proportion")
print(target_dist)

# -------------------------------------------------------
# 4.2 Define Leak-Free Features
# -------------------------------------------------------
numeric_features_clf = [
    col for col in [
        "pr_orderqty",
        "po_orderquantity",
        "po_netamount"
    ] if col in clf_df.columns
]

# Mandatory business dimensions (for stakeholder clarity)
categorical_features_clf = [
    col for col in [
        "pr_companycode",
        "po_purchasingorgdesc",
        "po_companycode",
        "po_companycodedesc",
        "pr_plant",
        "po_plant",
        "pr_documenttype",
        "po_purchasingdoctypedesc",
        "po_purchasinggroupdesc",
        "po_countrykey",
        "materialgroupdesc",
        "materialtypedesc",
        "record_type"
    ] if col in clf_df.columns
]

print("\n[Classification] Numeric features:", numeric_features_clf)
print("[Classification] Categorical features:", categorical_features_clf)

clf_model_df = clf_df[numeric_features_clf + categorical_features_clf + ["sla_breach_flag_bin"]].copy()

# Basic imputation
clf_model_df[numeric_features_clf] = clf_model_df[numeric_features_clf].fillna(0)
clf_model_df[categorical_features_clf] = clf_model_df[categorical_features_clf].fillna("Unknown")

X_clf = clf_model_df[numeric_features_clf + categorical_features_clf]
y_clf = clf_model_df["sla_breach_flag_bin"]

# -------------------------------------------------------
# 4.3 TrainTest Split
# -------------------------------------------------------
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X_clf,
    y_clf,
    test_size=0.20,
    random_state=42,
    stratify=y_clf
)

print("\nClassification train shape:", X_train_clf.shape)
print("Classification test shape :", X_test_clf.shape)

# -------------------------------------------------------
# 4.4 Preprocessing + SMOTE Balancing
# -------------------------------------------------------
preprocessor_clf = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features_clf),
        ("num", "passthrough", numeric_features_clf)
    ]
)

# Fit on train only
X_train_enc = preprocessor_clf.fit_transform(X_train_clf)
X_test_enc = preprocessor_clf.transform(X_test_clf)

# Apply SMOTE to balance minority class on the encoded features
sm = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = sm.fit_resample(X_train_enc, y_train_clf)

print("\nClass distribution before SMOTE:", dict(y_train_clf.value_counts()))
print("Class distribution after SMOTE :", dict(pd.Series(y_train_balanced).value_counts()))

# -------------------------------------------------------
# 4.5 Random Forest Classifier (balanced, threshold tuned)
# -------------------------------------------------------
rf_clf = RandomForestClassifier(
    n_estimators=120,
    max_depth=5,
    min_samples_leaf=10,
    class_weight="balanced",
    max_features="sqrt",
    random_state=42,
    bootstrap=True
)

rf_clf.fit(X_train_balanced, y_train_balanced)

# -------------------------------------------------------
# 4.6 Evaluate Classification Model + Export Predictions
# -------------------------------------------------------
# Probabilities from model
y_proba_clf = rf_clf.predict_proba(X_test_enc)[:, 1]

# Threshold tuned to avoid always predicting "YES"
threshold = 0.55
y_pred_clf = (y_proba_clf >= threshold).astype(int)

acc = accuracy_score(y_test_clf, y_pred_clf)
prec = precision_score(y_test_clf, y_pred_clf)
rec = recall_score(y_test_clf, y_pred_clf)
f1 = f1_score(y_test_clf, y_pred_clf)
auc = roc_auc_score(y_test_clf, y_proba_clf)

print("\n=== SLA Breach Classification (Balanced Model) ===")
print("Accuracy :", round(acc, 3))
print("Precision:", round(prec, 3))
print("Recall   :", round(rec, 3))
print("F1-score :", round(f1, 3))
print("ROC AUC  :", round(auc, 3))

# ---- Export classification predictions to Excel ----
clf_results = X_test_clf.copy().reset_index(drop=True)
clf_results["actual_sla_breach"] = y_test_clf.reset_index(drop=True)
clf_results["predicted_sla_breach"] = y_pred_clf
clf_results["predicted_sla_probability"] = y_proba_clf

clf_results["actual_sla_breach_label"] = clf_results["actual_sla_breach"].map({1: "YES", 0: "NO"})
clf_results["predicted_sla_breach_label"] = clf_results["predicted_sla_breach"].map({1: "YES", 0: "NO"})

clf_output_path = "outputs/sla_breach_predictions.xlsx"
clf_results.to_excel(clf_output_path, index=False)

print(f"\n Classification prediction table exported to: {clf_output_path}")
display(clf_results.head())

# ---- Confusion Matrix (PNG) ----
cm = confusion_matrix(y_test_clf, y_pred_clf)
print("\nConfusion Matrix:\n", cm)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.title("Confusion Matrix - SLA Breach Prediction (Balanced Model)")
plt.tight_layout()

cm_png_path = "outputs/sla_breach_confusion_matrix.png"
plt.savefig(cm_png_path, dpi=300)
plt.show()
print(f" Confusion matrix saved to: {cm_png_path}")

# ---- ROC Curve (PNG) ----
RocCurveDisplay.from_predictions(y_test_clf, y_proba_clf)
plt.title("ROC Curve - SLA Breach Prediction (Balanced Model)")
plt.tight_layout()

roc_png_path = "outputs/sla_breach_roc_curve.png"
plt.savefig(roc_png_path, dpi=300)
plt.show()
print(f" ROC curve saved to: {roc_png_path}")

# -------------------------------------------------------
# 4.7 Feature Importance  Classification (+ PNG)
# -------------------------------------------------------
rf_model_clf = rf_clf
ohe_clf = preprocessor_clf.named_transformers_["cat"]

encoded_cat_features = list(ohe_clf.get_feature_names_out(categorical_features_clf))
all_feature_names_clf = encoded_cat_features + numeric_features_clf

importances_clf = pd.Series(
    rf_model_clf.feature_importances_,
    index=all_feature_names_clf
).sort_values(ascending=False)

plt.figure(figsize=(10, 8))
importances_clf.head(20).plot(kind="barh", color='steelblue')
plt.gca().invert_yaxis()
plt.title("Top 20 Feature Importances  SLA Breach Classification Model")
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()

fi_clf_path = "outputs/sla_breach_feature_importance.png"
plt.savefig(fi_clf_path, dpi=300)
plt.show()
print(f" SLA feature importance chart saved to: {fi_clf_path}")

print("\nTop 20 features driving SLA breach prediction:")
display(importances_clf.head(20))


# =======================================================
# 5. REGRESSION MODEL  PRPO CYCLE TIME (Business Days)
# =======================================================

if "pr_to_po_ageing" not in df.columns:
    raise ValueError("pr_to_po_ageing column is missing in the dataset.")

reg_df = df[df["pr_to_po_ageing"].notna()].copy()
target_reg = "pr_to_po_ageing"

numeric_features_reg = [
    col for col in [
        "pr_approval_ageing",
        "po_approval_ageing",
        "pr_orderqty",
        "po_orderquantity",
        "po_netamount"
    ] if col in reg_df.columns
]

categorical_features_reg = [
    col for col in [
        "pr_companycode",
        "po_purchasingorgdesc",
        "po_companycode",
        "po_companycodedesc",
        "pr_plant",
        "po_plant",
        "pr_documenttype",
        "po_purchasingdoctype",
        "po_purchasingdoctypedesc",
        "po_purchasinggroupdesc",
        "po_countrykey",
        "materialgroupdesc",
        "materialtypedesc",
        "record_type"
    ] if col in reg_df.columns
]

print("\n[Regression] Numeric features:", numeric_features_reg)
print("[Regression] Categorical features:", categorical_features_reg)

reg_model_df = reg_df[numeric_features_reg + categorical_features_reg + [target_reg]].copy()

reg_model_df[numeric_features_reg] = reg_model_df[numeric_features_reg].fillna(0)
reg_model_df[categorical_features_reg] = reg_model_df[categorical_features_reg].fillna("Unknown")

X_reg = reg_model_df[numeric_features_reg + categorical_features_reg]
y_reg = reg_model_df[target_reg]

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg,
    y_reg,
    test_size=0.2,
    random_state=42
)

print("\nRegression train shape:", X_train_reg.shape)
print("Regression test shape :", X_test_reg.shape)

cat_transformer_reg = OneHotEncoder(handle_unknown="ignore")

preprocessor_reg = ColumnTransformer(
    transformers=[
        ("cat", cat_transformer_reg, categorical_features_reg),
        ("num", "passthrough", numeric_features_reg)
    ]
)

rf_reg = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

reg_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor_reg),
    ("model", rf_reg)
])

reg_pipeline.fit(X_train_reg, y_train_reg)

y_pred_reg = reg_pipeline.predict(X_test_reg)

mae = mean_absolute_error(y_test_reg, y_pred_reg)
mse = mean_squared_error(y_test_reg, y_pred_reg)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_reg, y_pred_reg)

print("\n=== PRPO Ageing Regression Performance ===")
print("MAE :", round(mae, 3), "business days")
print("RMSE:", round(rmse, 3), "business days")
print("R  :", round(r2, 3))

# ---- Actual vs Predicted Scatter (PNG) ----
plt.figure(figsize=(6, 6))
plt.scatter(y_test_reg, y_pred_reg, alpha=0.3)
plt.plot(
    [y_test_reg.min(), y_test_reg.max()],
    [y_test_reg.min(), y_test_reg.max()],
    "r--"
)
plt.xlabel("Actual PRPO Ageing (Business Days)")
plt.ylabel("Predicted PRPO Ageing (Business Days)")
plt.title("Actual vs Predicted PRPO Ageing")
plt.tight_layout()

scatter_png_path = "outputs/pr_to_po_ageing_scatter.png"
plt.savefig(scatter_png_path, dpi=300)
plt.show()
print(f" Regression scatter plot saved to: {scatter_png_path}")

# ---- Export regression predictions to Excel ----
reg_results = X_test_reg.copy().reset_index(drop=True)
reg_results["actual_pr_to_po_ageing"] = y_test_reg.reset_index(drop=True)
reg_results["predicted_pr_to_po_ageing"] = y_pred_reg
reg_results["error_days"] = reg_results["predicted_pr_to_po_ageing"] - reg_results["actual_pr_to_po_ageing"]

reg_output_path = "outputs/pr_to_po_ageing_predictions.xlsx"
reg_results.to_excel(reg_output_path, index=False)

print(f"\n Regression prediction table exported to: {reg_output_path}")
display(reg_results.head())

# -------------------------------------------------------
# 5.1 Feature Importance  Regression (+ PNG)
# -------------------------------------------------------
rf_model_reg = reg_pipeline.named_steps["model"]
ohe_reg = reg_pipeline.named_steps["preprocessor"].named_transformers_["cat"]

encoded_cat_features_reg = list(ohe_reg.get_feature_names_out(categorical_features_reg))
all_feature_names_reg = encoded_cat_features_reg + numeric_features_reg

importances_reg = pd.Series(
    rf_model_reg.feature_importances_,
    index=all_feature_names_reg
).sort_values(ascending=False)

plt.figure(figsize=(10, 8))
importances_reg.head(20).plot(kind="barh", color='darkorange')
plt.gca().invert_yaxis()
plt.title("Top 20 Feature Importances  PRPO Ageing Regression Model")
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()

fi_reg_path = "outputs/pr_to_po_feature_importance.png"
plt.savefig(fi_reg_path, dpi=300)
plt.show()
print(f" PRPO feature importance chart saved to: {fi_reg_path}")

print("\nTop 20 features driving PRPO cycle time:")
display(importances_reg.head(20))


# =======================================================
# 6. BUSINESS INSIGHTS  GROUPED ANALYSIS FOR STAKEHOLDERS
# =======================================================
print("\n=== Average PRPO Ageing by PO Company Code ===")
if "po_companycode" in df.columns:
    company_summary = (
        df[df["pr_to_po_ageing"].notna()]
        .groupby("po_companycode")["pr_to_po_ageing"]
        .mean()
        .sort_values(ascending=False)
    )
    display(company_summary)

print("\n=== Average PRPO Ageing by PO Plant (Top 10) ===")
if "po_plant" in df.columns:
    plant_summary = (
        df[df["pr_to_po_ageing"].notna()]
        .groupby("po_plant")["pr_to_po_ageing"]
        .mean()
        .sort_values(ascending=False)
        .head(10)
    )
    display(plant_summary)

print("\n=== Average PRPO Ageing by PO Document Type ===")
if "po_purchasingdoctypedesc" in df.columns:
    doc_type_summary = (
        df[df["pr_to_po_ageing"].notna()]
        .groupby("po_purchasingdoctypedesc")["pr_to_po_ageing"]
        .mean()
        .sort_values(ascending=False)
    )
    display(doc_type_summary)

print("\n=== Average PRPO Ageing by Vendor Country (Top 10) ===")
if "po_countrykey" in df.columns:
    country_summary = (
        df[df["pr_to_po_ageing"].notna()]
        .groupby("po_countrykey")["pr_to_po_ageing"]
        .mean()
        .sort_values(ascending=False)
        .head(10)
    )
    display(country_summary)

print("\n=== Script Complete  Models, Outputs & Insights Ready for Milestone 2 ===")
```
